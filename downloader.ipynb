```python
import yfinance as yf
import pandas as pd

tickers = {
    "CVX": "CVX",
    "XOM": "XOM"
}

for ticker, name in tickers.items():
    data = yf.download(
        ticker,
        period="max",
        interval="1d",
        auto_adjust=False,
        progress=False
    )

    filename = f"{name}_daily.csv"
    data.to_csv(filename, header=True)

    print(f"{ticker} data saved to {filename}")
```
