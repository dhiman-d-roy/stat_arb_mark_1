```python
import yfinance as yf
import pandas as pd
from pathlib import Path

DATA_DIR = Path("Data")
DATA_DIR.mkdir(exist_ok=True)

tickers = {
    "SPY": "SPY",
    "GLD": "GLD",
    "IAU": "IAU",
    "IVV": "IVV",
    "AGG": "AGG",
    "BND": "BND",
    "VTI": "VTI",
    "SCHB": "SCHB"
}

for ticker, name in tickers.items():
    data = yf.download(
        ticker,
        period="max",
        interval="1d",
        auto_adjust=False,
        progress=False
    )

    filename = DATA_DIR / f"{name}_daily.csv"
    data.to_csv(filename, header=True)

    print(f"{ticker} data saved to {filename}")
```
